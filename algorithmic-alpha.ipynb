```python
import pandas as pd

cashflow = pd.read_csv(r"C:\Users\Admin\Downloads\capital_allocation_project_datasets\data\raw\project_cash_flows.csv")
discount = pd.read_csv(r"C:\Users\Admin\Downloads\capital_allocation_project_datasets\data\raw\discount_rate_scenarios.csv")
sp100 = pd.read_csv(r"C:\Users\Admin\Downloads\capital_allocation_project_datasets\data\raw\sp100_company_data.csv")
stock = pd.read_csv(r"C:\Users\Admin\Downloads\capital_allocation_project_datasets\data\raw\stock_price_history_sample.csv")
macro = pd.read_csv(r"C:\Users\Admin\Downloads\capital_allocation_project_datasets\data\raw\macro_indicators_sample.csv")
wacc = pd.read_csv(r"C:\Users\Admin\Downloads\capital_allocation_project_datasets\data\raw\wacc_inputs.csv")
dcf = pd.read_csv(r"C:\Users\Admin\Downloads\capital_allocation_project_datasets\data\raw\dcf_forecast_assumptions.csv")
comps = pd.read_csv(r"C:\Users\Admin\Downloads\capital_allocation_project_datasets\data\raw\comparable_companies.csv")
transactions = pd.read_csv(r"C:\Users\Admin\Downloads\capital_allocation_project_datasets\data\raw\precedent_transactions.csv")
```

```python
sp100['pe_ratio'] = sp100['share_price_usd'] / sp100['eps']

sp100[['company','sector','pe_ratio']].head()
```

```python
sector_pe = (
    sp100.groupby('sector')['pe_ratio']
    .median()
    .sort_values(ascending=False)
)

print(sector_pe)
```

```python
import matplotlib.pyplot as plt

sector_pe.plot(kind='bar', figsize=(10,5))
plt.title('Median P/E Ratio by Sector')
plt.ylabel('P/E Ratio')
plt.show()
```

```python
volatility = (
    stock.groupby('ticker')['monthly_return']
    .std()
    .sort_values(ascending=False)
)

print(volatility)
```

```python
wacc['cost_of_equity'] = (
    wacc['risk_free_rate_pct']
    + wacc['beta']
    * wacc['equity_risk_premium_pct']
)

wacc['after_tax_cost_of_debt'] = (
    wacc['pre_tax_cost_of_debt_pct']
    * (1 - wacc['tax_rate_pct']/100)
)

wacc['wacc'] = (
    (wacc['equity_weight_pct']/100)*wacc['cost_of_equity']
    +
    (wacc['debt_weight_pct']/100)*wacc['after_tax_cost_of_debt']
)

wacc[['company','wacc']]
```
